```python
plt.figure(figsize=(13,8),dpi=150)
ax = sns.countplot(data=data, x='heart disease', palette="Set2")
for p in ax.patches:
    x=p.get_bbox().get_points()[:,0]
    y=p.get_bbox().get_points()[1,1]
    ax.annotate('{:.1f}%'.format(100.*y/len(data)), (x.mean(), y),
            ha='center', va='bottom')
```
