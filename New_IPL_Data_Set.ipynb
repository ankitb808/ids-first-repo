```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

```

```python
df1=pd.read_csv('/content/IPL Ball-by-Ball 2008-2020.csv')
df2=pd.read_csv('/content/IPL Matches 2008-2020.csv')
```

```python
df1
df2
```

```python
# Merge the two datasets on the 'id' column
merged_df = pd.merge(df1, df2, on='id')
merged_df
```

```python
# Save the merged dataset to a new CSV file
merged_df.to_csv('Merged_IPL_Dataset.csv', index=False)
```

```python
# Extract the year from the date column and add it as a new column 'season'
merged_df['season'] = pd.to_datetime(merged_df['date'], errors='coerce').dt.year
```

```python
# Group by 'season' and count the number of unique matches
matches_per_season = merged_df.groupby('season')['id'].nunique()
```

```python
# count of number of matches played in each season
matches_per_season
```

```python
# How many runs were scored in each season?
runs_per_season = merged_df.groupby('season')['batsman_runs'].sum()
```

```python
runs_per_season
```

```python
# What were the runs scored per match in different seasons?
runs_per_match = runs_per_season / matches_per_season
```

```python
runs_per_match
```
