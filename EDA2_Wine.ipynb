```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv('/content/Wine.csv')
```

```python
# Rename the column
df.rename(columns={"OD280/OD315 of diluted wines": "Quality"}, inplace=True)
```

```python
# visualize the distribution
sns.distplot(df['Alcohol'])
```

```python
# Correlation matrix
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# pairplot to explore the relationships
sns.pairplot(df)
```

```python
# Boxplot for feature distributions by quality
plt.figure(figsize=(10, 6))
sns.boxplot(x='Quality', y='Alcohol', data=df)
plt.title('Alcohol Content by Quality')
plt.show()
```

1.The correlation matrix shows significant relationships between various features and wine quality
2.The pairplot suggests that some features might have non-linear relationships with the target variable
3.The boxplot indicates that higher quality wines tend to have higher alcohol content.
